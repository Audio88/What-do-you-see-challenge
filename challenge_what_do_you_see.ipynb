```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = pd.read_sas('hads2013n.sas7bdat')
df[['LMED', 'APLMED', 'FMR', 'FMTBUILT', 'ROOMS']]
%matplotlib inline
```

```python
house_income = df['LMED']
house_income = house_income.dropna()

house_income_adj = df['APLMED']
house_income_adj = house_income_adj.dropna()

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Before')
plt.boxplot(house_income)
plt.ylabel('Income')

plt.subplot(1, 2, 2)
plt.title('After')
plt.boxplot(house_income_adj)
plt.ylabel('Income')

plt.tight_layout()
plt.show()
```

Two boxplots showing income before and after being adjusted by standard HUD adjustment factors. On average incomes come down, but a few outliers earn more.

```python
plt.hist(df['FMR'])
plt.ylabel('Number of People')
plt.xlabel('Cost of Rent')
plt.show()
```

Majority of people pay 750\$ to 1600\$ in rent per month.

```python
year_built = df['FMTBUILT']
fig, ax = plt.subplots()
plt.hist(year_built, bins=7)
plt.ylabel('Number of Houses')
plt.xlabel('Year built')
fig.autofmt_xdate()
plt.show()
```

Majority of the homes were built before 1980.

```python
average_rent = df.groupby(['FMR'], as_index=False).mean()

plt.scatter(y=average_rent['FMR'], x=average_rent['ROOMS'])
plt.ylabel('Cost of Rent')
plt.xlabel('Number of Rooms')
plt.show()
```

Here you can see the relation between number of rooms and cost of rent.
